```python
selected_columns = ['Id', 'BedroomAbvGr', 'FullBath', 'HalfBath', '1stFlrSF', '2ndFlrSF', 'TotalBsmtSF', 'GrLivArea']
cleanedTrain = trainset[selected_columns].copy()
cleanedTrain.head()
```

```python
cleanedTrain['SquareFootage'] = cleanedTrain['1stFlrSF'] + cleanedTrain['2ndFlrSF'] + cleanedTrain['TotalBsmtSF'] + cleanedTrain['GrLivArea']
cleanedTrain['TotalRooms'] = cleanedTrain['BedroomAbvGr'] + cleanedTrain['FullBath'] + cleanedTrain['HalfBath']
cleanedTrain.head()
```

```python
cleanedTrain = cleanedTrain[['Id', 'SquareFootage', 'TotalRooms']]
cleanedTrain.head()
```
